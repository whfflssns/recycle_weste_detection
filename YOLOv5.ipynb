```python
## 이미지 크기와 boundbox 얻는 코드
## box 에는 파일 순서, 왼쪽상단 x,y 좌표, 오른쪽하단 x,y좌표, width, height, 라벨으로 되어있고
## size는 파일 순서, 크기 이렇게되어있다.
from PIL import Image

size_a = []
box = []
for file in os.listdir(train_dir):
  if file.endswith('.roi'):
    x = read_roi_file(train_dir+file)
    a1 = x[file.split('.')[0]]
    box.append([file.split('_')[0], a1['left'], a1['top'], a1['left']+a1['width'], a1['top']+a1['height'], a1['width'], a1['height'] ,a1['name'].split('_')[2]])
  elif file.endswith('.jpg'):
    a = Image.open(train_dir+file)
    w, h = a.size
    size_a.append([file.split('.')[0], w, h])

box.sort()
size_a.sort()
print(box[0], size_a[0])
```

```python
## 임의로 오류를 만들어 문제가 생긴다면 처리할 수 있도록 만들 예외
class MyError(Exception):
  pass
```

```python
## 우리가 사용하던 불편한 roi데이터를 YOLO format으로 대리고 올 convert함수

def convert(size, box):
  cl = -1
  dw = 1./size[1] 
  dh = 1./size[2] 
  x = (box[1] + box[3])/2.0 
  y = (box[2] + box[4])/2.0 
  w = box[5]
  h = box[6]
  x = x*dw 
  w = w*dw 
  y = y*dh 
  h = h*dh 
  if box[-1] == 'pl':
    cl = 0
  elif box[-1] == 'pp':
    cl = 1
  elif box[-1] == 'mt':
    cl = 2
  elif box[-1] == 'gl':
    cl = 3
  elif box[-1] == 'nr':
    cl = 4
  if cl == -1:
    raise MyError()
  return [cl,x,y,w,h]
```

```python
for i in range(len(size_a)):
  aaa = []
  for j in range(len(box)):
    if size_a[i][0] == box[j][0]:
      aaa.append(convert(size_a[i], box[j]))
  with open(train_dir+size_a[i][0]+'.txt','w', newline='') as f:
      for i in range(len(aaa)):
        f.write(str(aaa[i][0])+' '+str(aaa[i][1])+' '+str(aaa[i][2])+' '+str(aaa[i][3])+' '+str(aaa[i][4])+'\n')
```
